```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from scipy.integrate import solve_ivp

# ============================================================
# Part 1: Snake-skin friction wrench for one body element
# ============================================================

def snake_skin_wrench(Vb, f_parallel=1.0, f_perp=2.0):
    """
    Input:
        Vb = [omega, vx, vy]
    
    Output:
        Fb = [torque, Fx, Fy]
    """

    omega, vx, vy = Vb

    # Friction in body x-direction
    if vx < 0:
        Fx = -2 * f_parallel * np.sign(vx)   # sliding backward
    elif vx > 0:
        Fx = -1 * f_parallel * np.sign(vx)   # sliding forward
    else:
        Fx = 0.0

    # Friction in body y-direction
    if vy != 0:
        Fy = -f_perp * np.sign(vy)
    else:
        Fy = 0.0

    torque = 0.0

    return np.array([torque, Fx, Fy])


# ============================================================
# Helper functions
# ============================================================

def R(theta):
    """
    2D rotation matrix
    """
    return np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta),  np.cos(theta)]
    ])


def adjoint_SE2(theta, p):
    """
    Planar adjoint matrix for twist order:
        V = [omega, vx, vy]
    """

    Rot = R(theta)

    Ad = np.array([
        [1.0,      0.0,        0.0],
        [p[1],     Rot[0, 0],  Rot[0, 1]],
        [-p[0],    Rot[1, 0],  Rot[1, 1]]
    ])

    return Ad


# ============================================================
# Part 2: Net snake-skin friction for three-link swimmer
# ============================================================

def net_snake_friction(q, qdot, l=1.0, f_parallel=1.0, f_perp=2.0):
    """
    q = [x, y, theta, alpha1, alpha2]
    qdot = [xdot, ydot, thetadot, alpha1dot, alpha2dot]

    Returns:
        F_net = [net torque, net Fx, net Fy]
    """

    x, y, theta, alpha1, alpha2 = q
    xdot, ydot, thetadot, alpha1dot, alpha2dot = qdot

    # Middle link body twist
    V_middle = np.array([thetadot, xdot, ydot])
    F_middle = snake_skin_wrench(V_middle, f_parallel, f_perp)

    # Front link
    p_front = np.array([l, 0.0])
    Ad_front = adjoint_SE2(alpha1, p_front)

    V_front = Ad_front @ V_middle + np.array([alpha1dot, 0.0, 0.0])
    F_front_local = snake_skin_wrench(V_front, f_parallel, f_perp)

    # Convert front wrench to middle frame
    F_front = Ad_front.T @ F_front_local

    # Back link
    p_back = np.array([-l, 0.0])
    Ad_back = adjoint_SE2(alpha2, p_back)

    V_back = Ad_back @ V_middle + np.array([alpha2dot, 0.0, 0.0])
    F_back_local = snake_skin_wrench(V_back, f_parallel, f_perp)

    # Convert back wrench to middle frame
    F_back = Ad_back.T @ F_back_local

    # Total wrench
    F_net = F_middle + F_front + F_back

    return F_net


# ============================================================
# Part 3: Use fsolve for one instant in time
# ============================================================

q_random = np.array([
    0.0,    # x
    0.0,    # y
    0.3,    # theta
    0.5,    # alpha1
   -0.4     # alpha2
])

alpha1dot_random = 1.2
alpha2dot_random = -0.8

def force_balance_one_instant(V_unknown):
    """
    V_unknown = [xdot, ydot, thetadot]
    """

    xdot, ydot, thetadot = V_unknown

    qdot = np.array([
        xdot,
        ydot,
        thetadot,
        alpha1dot_random,
        alpha2dot_random
    ])

    return net_snake_friction(q_random, qdot)


V_guess = np.array([0.0, 0.0, 0.0])
V_solution = fsolve(force_balance_one_instant, V_guess)

print("Part 3 solution:")
print("xdot     =", V_solution[0])
print("ydot     =", V_solution[1])
print("thetadot =", V_solution[2])

print("\nForce balance check:")
print(force_balance_one_instant(V_solution))


# ============================================================
# Part 4: Integrate locomotion using solve_ivp
# ============================================================

A = np.pi / 4
omega_gait = 2 * np.pi
phase = np.pi / 2

l = 1.0
f_parallel = 1.0
f_perp = 2.0


def gait_angles(t):
    alpha1 = A * np.sin(omega_gait * t)
    alpha2 = A * np.sin(omega_gait * t + phase)

    return alpha1, alpha2


def gait_rates(t):
    alpha1dot = A * omega_gait * np.cos(omega_gait * t)
    alpha2dot = A * omega_gait * np.cos(omega_gait * t + phase)

    return alpha1dot, alpha2dot


def snake_dynamics(t, state):
    """
    state = [x, y, theta]

    Returns:
        [xdot, ydot, thetadot]
    """

    x, y, theta = state

    alpha1, alpha2 = gait_angles(t)
    alpha1dot, alpha2dot = gait_rates(t)

    q = np.array([x, y, theta, alpha1, alpha2])

    def force_balance(V_unknown):
        xdot, ydot, thetadot = V_unknown

        qdot = np.array([
            xdot,
            ydot,
            thetadot,
            alpha1dot,
            alpha2dot
        ])

        return net_snake_friction(
            q,
            qdot,
            l=l,
            f_parallel=f_parallel,
            f_perp=f_perp
        )

    V_guess = np.array([0.0, 0.0, 0.0])
    V_solution = fsolve(force_balance, V_guess)

    return V_solution


# Initial condition
state0 = np.array([
    0.0,    # x
    0.0,    # y
    0.0     # theta
])

# Simulate for 5 gait cycles
t_span = (0, 5)
t_eval = np.linspace(t_span[0], t_span[1], 1000)

sol = solve_ivp(
    snake_dynamics,
    t_span,
    state0,
    t_eval=t_eval,
    rtol=1e-6,
    atol=1e-8
)


# ============================================================
# Plots
# ============================================================

x_vals = sol.y[0]
y_vals = sol.y[1]
theta_vals = sol.y[2]

plt.figure(figsize=(6, 6))
plt.plot(x_vals, y_vals)
plt.xlabel("x position")
plt.ylabel("y position")
plt.title("Three-Link Snake Locomotion Trajectory")
plt.axis("equal")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 4))
plt.plot(sol.t, theta_vals)
plt.xlabel("Time")
plt.ylabel(r"$\theta$")
plt.title("Snake Orientation Over Time")
plt.grid(True)
plt.show()

alpha1_vals = []
alpha2_vals = []

for t in sol.t:
    a1, a2 = gait_angles(t)
    alpha1_vals.append(a1)
    alpha2_vals.append(a2)

plt.figure(figsize=(8, 4))
plt.plot(sol.t, alpha1_vals, label=r"$\alpha_1$")
plt.plot(sol.t, alpha2_vals, label=r"$\alpha_2$")
plt.xlabel("Time")
plt.ylabel("Joint Angle")
plt.title("Sinusoidal Joint Gait")
plt.legend()
plt.grid(True)
plt.show()
```
